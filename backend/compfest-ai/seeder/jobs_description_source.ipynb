```python
import pandas as pd
EXTRACT_FEATURE_MAP = [
    "Experience",
    "Qualifications",
    "Salary Range",
    "Work Type",
    "Job Title",
    "Role",
    "Job Description",
    "skills",
    "Responsibilities"
]

def load_data(path: str, debug: bool):
    df = pd.read_csv(path, on_bad_lines='skip', nrows=100000)
    if debug:
        print(f"Original number of rows: {df.shape[0]}")
        print(f"Columns in the dataframe: {df.columns}")
    return df

```

```python
mkdir /kaggle/working/cleaned/
```

```python
DATA_PATH = "/kaggle/input/job-description-dataset/job_descriptions.csv"
OUTPUT_PATH = '/kaggle/working/cleaned/job_descriptions.csv'

df = load_data(DATA_PATH, debug=False)
df_selected = df[EXTRACT_FEATURE_MAP]

df_unique = df.drop_duplicates(subset=['Job Title'])
df_unique.columns = df_unique.columns.str.lower().str.replace(' ', '_')


df_unique.to_csv(OUTPUT_PATH, index=False)
```
